It search every local news tweet handle that mentioning the keyword of the crime as for examble bunuh ( stands for murder/kill in Bahasa Malay)

```python
import pandas as pd
import numpy as np


df = pd.read_csv("E:/Notebook/Twitter/samun/samun.csv")
dftemp=pd.read_csv("E:/Notebook/Twitter/canvas.csv")

majmuk=['dua','tiga','empat','lima','enam','tujuh','lapan','sembilan','sepuluh'
        ,'puluh','belas','kumpulan','geng','kawan','pasangan']
singular=['seorang','satu suspek']
senjataapi=['pistol','senjata api','senapang','tembak']

for i in df.index:
    text=df.at[i, 'tweet']
    print(text)
    
    if any(a in text for a in majmuk):
        if any(a in text for a in senjataapi):
            dftemp.at[i, 'text']=text
            dftemp.at[i, 'jenayah']="samun berkumpulan bersenjata api"
        else:
            dftemp.at[i, 'text']=text
            dftemp.at[i, 'jenayah']="samun berkumpulan tanpa senjata api"        
    else:
        if any(a in text for a in senjataapi):
            dftemp.at[i, 'text']=text
            dftemp.at[i, 'jenayah']="samun seorang bersenjata api"
        else:
            dftemp.at[i, 'text']=text
            dftemp.at[i, 'jenayah']="samun seorang tidak senjata api"
    

dftemp['text'].replace('', np.nan, inplace=True)
dftemp.dropna(subset=['text'], inplace=True)
dftemp.to_csv('E:/Notebook/MODEL/Samun.csv', mode='a', header=False,index=False)
print('success')
```
