```python
#functional selected model singlemodel

import pickle





def validationconfidence (model,text,classes):
    predictionConfidence = model.predict_proba(text)
#     print("The Confidance scores:", predictionConfidence)
    prediction =model.predict(text)
    maxConfidence = max(predictionConfidence[0])
   

    if maxConfidence > 0.8:
#         print(prediction)
        
        result=classes[int(prediction)]

        return (result,maxConfidence)
    else:
        result=classes[int(prediction)]
        result='Memerlukan siasatan terperinci'
        return (result,maxConfidence)

def vectorizingstring(string):
  import pickle
  with open('/content/drive/MyDrive/Colab Notebooks/Data/newvector.pickel', 'rb') as handle:
    vectorizer = pickle.load(handle)
  vectorizedtext=vectorizer.transform([string])

  return vectorizedtext

def loadtokenizer():
  import pickle
  with open('/content/drive/MyDrive/Colab Notebooks/Data/tokenizerrevisedCNN.pickle', 'rb') as handle:
    tokenizer = pickle.load(handle)
  return tokenizer

def cnnstringformat(string,tokenizer):
  from keras.preprocessing.text import Tokenizer
  from keras.preprocessing.sequence import pad_sequences
  import pandas as pd
  MAX_SEQUENCE_LENGTH = 1000
  instance = tokenizer.texts_to_sequences(string)
  flat_list = []
  for sublist in instance:
      for item in sublist:
          flat_list.append(item)
  flat_list = [flat_list]
  instance = pad_sequences(flat_list, maxlen=MAX_SEQUENCE_LENGTH)

  return instance

def validationconfidenceforCNN (model,text,classes):
    predictionConfidence = model.predict(text) 
#     print("The Confidance scores:", predictionConfidence)
    prediction =predictionConfidence.argmax(axis=-1)
    maxConfidence = max(predictionConfidence[0])
    
    if maxConfidence > 0.8:
#         print(prediction)
        
        result=classes[int(prediction)]
        return (result,maxConfidence)
    else:
        result=classes[int(prediction)]
        result='Memerlukan siasatan terperinci'
        return (result,maxConfidence)



def computeresulttomodel(model,string,classes):
  
  text=vectorizingstring(string)

  
  kelastekaan_model,kebarangkalian_model=validationconfidence(model,text,classes)
  
  return (kelastekaan_model,kebarangkalian_model)


def computeresulttocnnmodel(model,string,classes):
  
  text=cnnstringformat(string,loadtokenizer())

  
  kelastekaan_model,kebarangkalian_model=validationconfidence(model,text,classes)
  
  return (kelastekaan_model,kebarangkalian_model)






def predict_text_svm (text):
  # load svm model
  text
  import pickle
  model = pickle.load(open('/content/drive/MyDrive/Colab Notebooks/Model/hypsvm3.sav', 'rb'))
  import pandas as pd
  classes=['Bunuh', 'Cabul Kehormatan', 'Mencederakan', 'Merusuh', 'Pecah Rumah dan Curi', 'Pemerasan', 'Ragut', 'Rogol', 'Ugutan Jenayah', 'curi', 'curi kenderaan berat', 'curi motokar', 'samun berkumpulan bersenjata api', 'samun berkumpulan tanpa senjata api', 'samun seorang bersenjata api', 'samun seorang tidak senjata api']

  kelasjenayah,kebarangkalian=computeresulttomodel(model,text,classes)

  
  round((kebarangkalian*100),4)
  return (kelasjenayah,round((kebarangkalian*100),4))

def predict_text_nb (text):
  # load svm model
  text
  import pickle
  model = pickle.load(open('/content/drive/MyDrive/Colab Notebooks/Model/BaselinedNb.sav', 'rb'))
  import pandas as pd
  classes=['Bunuh', 'Cabul Kehormatan', 'Mencederakan', 'Merusuh', 'Pecah Rumah dan Curi', 'Pemerasan', 'Ragut', 'Rogol', 'Ugutan Jenayah', 'curi', 'curi kenderaan berat', 'curi motokar', 'samun berkumpulan bersenjata api', 'samun berkumpulan tanpa senjata api', 'samun seorang bersenjata api', 'samun seorang tidak senjata api']

  kelasjenayah,kebarangkalian=computeresulttomodel(model,text,classes)

  
  round((kebarangkalian*100),4)
  return (kelasjenayah,round((kebarangkalian*100),4))

def predict_text_lr (text):
  # load svm model
  text
  import pickle
  model = pickle.load(open('/content/drive/MyDrive/Colab Notebooks/Model/hyplr.sav', 'rb'))
  import pandas as pd
  classes=['Bunuh', 'Cabul Kehormatan', 'Mencederakan', 'Merusuh', 'Pecah Rumah dan Curi', 'Pemerasan', 'Ragut', 'Rogol', 'Ugutan Jenayah', 'curi', 'curi kenderaan berat', 'curi motokar', 'samun berkumpulan bersenjata api', 'samun berkumpulan tanpa senjata api', 'samun seorang bersenjata api', 'samun seorang tidak senjata api']

  kelasjenayah,kebarangkalian=computeresulttomodel(model,text,classes)

  
  round((kebarangkalian*100),4)
  return (kelasjenayah,round((kebarangkalian*100),4))

def predict_text_cnn (text):
  # load svm model
  text
  import pickle
  from keras.models import load_model
  cnn = load_model('/content/drive/MyDrive/Colab Notebooks/Model/CNNhypertunedV2.hdf5')
  import pandas as pd
  classes=['Bunuh', 'Cabul Kehormatan', 'Mencederakan', 'Merusuh', 'Pecah Rumah dan Curi', 'Pemerasan', 'Ragut', 'Rogol', 'Ugutan Jenayah', 'curi', 'curi kenderaan berat', 'curi motokar', 'samun berkumpulan bersenjata api', 'samun berkumpulan tanpa senjata api', 'samun seorang bersenjata api', 'samun seorang tidak senjata api']
  instance=cnnstringformat(text,loadtokenizer())
  kelasjenayah,kebarangkalian=validationconfidenceforCNN(cnn,instance,classes)

  
  round((kebarangkalian*100),4)
  return (kelasjenayah,round((kebarangkalian*100),4))


# string='Saya telah menjumpai mayat di belakang semak'
# import sys
# sys.path.insert(0,'/content/drive/MyDrive/Colab Notebooks')

# from preprocesscleaningtext import preprocess


# text=preprocess(string)

# kelasjenayah,kebarangkalian=predict_text_nb(text)
# print('nb')
# print('Hasil Tekaan Jenayah:',kelasjenayah)
# print('Peratus Keyakinan:',kebarangkalian)

# kelasjenayah,kebarangkalian=predict_text_svm(text)
# print('svm')
# print('Hasil Tekaan Jenayah:',kelasjenayah)
# print('Peratus Keyakinan:',kebarangkalian)

# kelasjenayah,kebarangkalian=predict_text_lr(text)
# print('lr')
# print('Hasil Tekaan Jenayah:',kelasjenayah)
# print('Peratus Keyakinan:',kebarangkalian)

# kelasjenayah,kebarangkalian=predict_text_cnn(text)
# print('cnn')
# print('Hasil Tekaan Jenayah:',kelasjenayah)
# print('Peratus Keyakinan:',kebarangkalian)
```
