```python
import re
import malaya

corrector = malaya.spell.probability()
normalizer = malaya.normalize.normalizer(corrector)

def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
#remove url
def removeurl(text):
    regex='(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:\'".,<>?«»“”‘’]))'
    URLless_string = re.sub(r regex, '', text)
    return (URLless_string)

#remove special characters
def removespecial(text):
    text=re.sub(r"[^a-zA-Z0-9]+", ' ',text)
    return (text)

#remove numbers and alpha numeric
def removenumerical(text):
    only_a = []
      for word in text.split():
            if not re.search('\d',word):
                only_a.append(word)
          
        text=' '.join(only_a)

    return (text)

def translatestring(text):
    
  # text=transformer.greedy_decoder([text])
  # text=' '.join(text)
    text=transformer_large.greedy_decoder([text])
    text=' '.join(text)
    
    return text
  
def stemmingstring(text):
    
    text=model.stem(text)
    
    return (text)

def removemalaystopword(text):
    stop_words = open('/content/drive/MyDrive/Colab Notebooks/Preprocess/stopword (1).txt','r').read().split()
    storedlist=[] 
      # print(stop_words)

    a=text.split()
     for r in a: 
        if not r in stop_words:
          storedlist.append(r)

        if not storedlist:
            return ("Error - text contain no meaningful words")
        else:
            text=' '.join(storedlist)
    
    return (text)

# additional preprocessing for previous model development
# def removehashtagicopnet(text):
#   word='#icopnet'
#   if not word in text:
#     return (text)
#   else:
#     text = text.replace('#icopnet', '')
#     return (text)

def lowercase(text):
    
    text=text.lower()
    
    return (text)

import re,string

def strip_links(text):
    link_regex    = re.compile('((https?):((//)|(\\\\))+([\w\d:#@%/;$()~_?\+-=\\\.&](#!)?)*)', re.DOTALL)
    links         = re.findall(link_regex, text)
    for link in links:
        text = text.replace(link[0], ', ')    
    return text

def strip_all_entities(text):
    entity_prefixes = ['@','#']
    for separator in  string.punctuation:
        if separator not in entity_prefixes :
            text = text.replace(separator,' ')
    words = []
    for word in text.split():
        word = word.strip()
        if word:
            if word[0] not in entity_prefixes:
                words.append(word)
    return ' '.join(words)

def normalize(text):
    text=normalizer.normalize(text)
    a=text['normalize']
    
    return(a)

def detectlang(text):
    from langdetect import detect
    lang = detect(text)
    
    return(lang)


        
```
